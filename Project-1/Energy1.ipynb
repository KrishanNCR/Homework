```python
# Dependencies
%matplotlib notebook
# Import Dependencies
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import pandas as pd
```

```python
file = 'Total.csv'
df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
df.head()
```

```python
dfc = df.rename(columns={'Total Energy Production, trillion Btu': 'Production Energy','Total Energy Consumption, million Btu':'Consumption Energy','Average Retail Price of Electricity to Residential Sector, cents/kWh':'Retail Price of Electricity'})

```

```python
dfc.head()
```

```python
plt.figure(figsize=(10,5))

#img = mpimg.imread('power-grid-banner.jpg')
#imgplot = plt.imshow(img,zorder=-1, extent=[0.5, 8.0, 1.0, 7.0])
width = 0.5
ind = np.arange(len(df['State']))
tick_locations = [value+0.1 for value in ind]
leg =["Consumption(BTU)", "Production (BTU)"]

p1 = plt.bar(ind, df['Total Energy Production, trillion Btu'], width,color= "blue", edgecolor='black',alpha=0.5, align="edge",zorder=2)
p2 = plt.bar(ind + .5, df['Total Energy Consumption, million Btu'], width,color=[1, 0.2, 0.2], edgecolor='black',alpha=0.5, align="edge",zorder=2)
# Create a legend for our chart
plt.legend(leg, loc="best")
plt.xticks(tick_locations, df["State"], rotation="vertical")
plt.annotate('Maximum Production', xy=(43.5,36000), xytext=(50,40000),arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Minimum Production', xy=(8,500), xytext=(8,20000),arrowprops=dict(facecolor='black', shrink=1000))

#imgplot = plt.imshow(img,zorder= -1)
# Show the chart

plt.show()

```
